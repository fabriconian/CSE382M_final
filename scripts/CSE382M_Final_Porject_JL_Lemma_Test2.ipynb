```python
from PIL import Image
import os

input_folder = 'C:/Users/karlf/Downloads/archive/flowers'
output_folder = 'C:/Users/karlf/Downloads/output/flowers'
size = (64, 64)  # specify the size you want

# loop through each category folder
for category in os.listdir(input_folder):
    category_path = os.path.join(input_folder, category)
    if os.path.isdir(category_path):
        # loop through each image file in the category folder
        for file in os.listdir(category_path):
            file_path = os.path.join(category_path, file)
            if os.path.isfile(file_path):
                # open the image file and resize it
                with Image.open(file_path) as image:
                    resized_image = image.resize(size)
                    # save the resized image to the output folder
                    output_path = os.path.join(output_folder, category)
                    os.makedirs(output_path, exist_ok=True)
                    output_file_path = os.path.join(output_path, file)
                    resized_image.save(output_file_path)
```

```python
#change ourput images to RGB array form
import numpy as np
#loop through each category folder in output folder
lis_images = []
for category in os.listdir(output_folder):
    category_path_new = os.path.join(output_folder, category)
    lis_images_single_category = []
    if os.path.isdir(category_path_new):
        # loop through each image file in the category folder
        for file in os.listdir(category_path_new):
            file_path = os.path.join(category_path_new, file)
            #change the image in output folder to RGB array form
            if os.path.isfile(file_path):
                image_here = Image.open(file_path) 
                array_im = np.asarray(image_here)
                lis_images_single_category.append(array_im)
    lis_images.append(lis_images_single_category)
            
    
```

```python
#reshape images in RGB array form to long single vector form
#we'll call these image data in single vector form image vectors in following code image vector
#each of number in the image vector is an RGB value for original image
lis_image_vectors = []
for i in range(0,len(lis_images)):
    lis_image_vectors_category = []
    for j in range(0,len(lis_images[i])):
        image_shape = lis_images[i][j].shape
        lis_image_vectors_category.append(lis_images[i][j].reshape((image_shape[0]*image_shape[1]*image_shape[2],)))
    lis_image_vectors.append(lis_image_vectors_category)
```

```python
#generating single image vector data for 5 datasets of daisy, dandelion, rose, sunflower and tulip.
import math
#Test JL Lemma here
#Let c in JL Lemma be 10. k be 100.
c = 10
epsilon = 0.1

#length of a single image vector
len_image_vector_here=len(lis_image_vectors[0][0])

#flower data for different categories of flowers
daisy_data = np.array(lis_image_vectors[0])
dandelion_data = np.array(lis_image_vectors[1])
rose_data = np.array(lis_image_vectors[2])
sunflower_data = np.array(lis_image_vectors[3])
tulip_data = np.array(lis_image_vectors[4])
```

```python
import pandas as pd
#Generating and storing norms of difference between
#pairs of image vectors in five dataset of image vectors generated in cell above.
daisy_collection = np.zeros((len(daisy_data),len(daisy_data)))
dandelion_collection = np.zeros((len(dandelion_data),len(dandelion_data)))
rose_collection = np.zeros((len(rose_data),len(rose_data)))
sunflower_collection = np.zeros((len(sunflower_data),len(sunflower_data)))
tulip_collection = np.zeros((len(tulip_data),len(tulip_data)))
#compute norm of difference between
#pairs of image vectors in five dataset of image vectors
l1 = len(daisy_collection)
l2 = len(dandelion_collection)
l3 = len(rose_collection)
l4 = len(sunflower_collection)
l5 = len(tulip_collection)
for i in range(0,l1):
    for j in range(0,l1):
        v_i = np.array(daisy_data[i],dtype='float64')
        v_j = np.array(daisy_data[j],dtype='float64')
        norm_dif = np.linalg.norm(v_i-v_j)
        daisy_collection[i,j]=norm_dif
for i in range(0,l2):
    for j in range(0,l2):
        v_i = np.array(dandelion_data[i],dtype='float64')
        v_j = np.array(dandelion_data[j],dtype='float64')
        norm_dif = np.linalg.norm(v_i-v_j)
        dandelion_collection[i,j]=norm_dif
for i in range(0,l3):
    for j in range(0,l3):
        v_i = np.array(rose_data[i],dtype='float64')
        v_j = np.array(rose_data[j],dtype='float64')
        norm_dif = np.linalg.norm(v_i-v_j)
        rose_collection[i,j]=norm_dif
for i in range(0,l4):
    for j in range(0,l4):
        v_i = np.array(sunflower_data[i],dtype='float64')
        v_j = np.array(sunflower_data[j],dtype='float64')
        norm_dif = np.linalg.norm(v_i-v_j)
        sunflower_collection[i,j]=norm_dif
for i in range(0,l5):
    for j in range(0,l5):
        v_i = np.array(tulip_data[i],dtype='float64')
        v_j = np.array(tulip_data[j],dtype='float64')
        norm_dif = np.linalg.norm(v_i-v_j)
        tulip_collection[i,j]=norm_dif
```

```python
#Now we compute norms after projction
daisy_norm_before_proj = pd.read_csv('daisy_before_proj.csv')
dandelion_norm_before_proj = pd.read_csv('dandelion_before_proj.csv')
rose_norm_before_proj = pd.read_csv('rose_before_proj.csv')
sunflower_norm_before_proj = pd.read_csv('sunflower_before_proj.csv')
tulip_norm_before_proj = pd.read_csv('tulip_before_proj.csv')

```

```python
from sklearn.random_projection import johnson_lindenstrauss_min_dim, GaussianRandomProjection
eps = 0.1
#compute k values separately for daisy data, dandelion data, rose data, sunflower data and tulip data
n_samples1, n_features1 = daisy_data.shape
k_daisy1 = johnson_lindenstrauss_min_dim(n_samples1, eps=eps)
k_daisy2 = math.ceil(k_daisy1 / 2)
k_daisy3 = math.ceil(k_daisy1/4)



#compute and record record norm of differences between image vectors for daisy data after projection
rp_daisy1 = GaussianRandomProjection(n_components=k_daisy1, random_state=42)
daisy_rp1 = rp_daisy1.fit_transform(daisy_data)
rp_daisy2 = GaussianRandomProjection(n_components=k_daisy2, random_state=42)
daisy_rp2 = rp_daisy2.fit_transform(daisy_data)
rp_daisy3 = GaussianRandomProjection(n_components=k_daisy3, random_state=42)
daisy_rp3 = rp_daisy3.fit_transform(daisy_data)

daisy_norm_df1 = np.zeros((len(daisy_data),len(daisy_data)))
daisy_norm_df2 = np.zeros((len(daisy_data),len(daisy_data)))
daisy_norm_df3 = np.zeros((len(daisy_data),len(daisy_data)))
for i in range(0,len(daisy_data)):
    for j in range(0,len(daisy_data)):
        norm_dif = np.linalg.norm(daisy_rp1[i]-daisy_rp1[j])
        daisy_norm_df1[i,j]=norm_dif
for i in range(0,len(daisy_data)):
    for j in range(0,len(daisy_data)):
        norm_dif = np.linalg.norm(daisy_rp2[i]-daisy_rp2[j])
        daisy_norm_df2[i,j]=norm_dif
for i in range(0,len(daisy_data)):
    for j in range(0,len(daisy_data)):
        norm_dif = np.linalg.norm(daisy_rp3[i]-daisy_rp3[j])
        daisy_norm_df3[i,j]=norm_dif
        
name1 = 'daisy_after_proj '+ str(k_daisy1) +'.csv'
name2 = 'daisy_after_proj '+ str(k_daisy2) +'.csv'
name3 = 'daisy_after_proj '+ str(k_daisy3) + '.csv'

pd.DataFrame(daisy_norm_df1).to_csv(name1)
pd.DataFrame(daisy_norm_df2).to_csv(name2)
pd.DataFrame(daisy_norm_df3).to_csv(name3)

#case of dandelion data after projection
n_samples2, n_features2 = dandelion_data.shape
k_dandelion1 = johnson_lindenstrauss_min_dim(n_samples2, eps=eps)
k_dandelion2 = math.ceil(k_dandelion1 / 2)
k_dandelion3 = math.ceil(k_dandelion1/4)



#compute and record record norm of differences between image vectors for dandelion data after projection
rp_dandelion1 = GaussianRandomProjection(n_components=k_dandelion1, random_state=42)
dandelion_rp1 = rp_dandelion1.fit_transform(dandelion_data)
rp_dandelion2 = GaussianRandomProjection(n_components=k_dandelion2, random_state=42)
dandelion_rp2 = rp_dandelion2.fit_transform(dandelion_data)
rp_dandelion3 = GaussianRandomProjection(n_components=k_dandelion3, random_state=42)
dandelion_rp3 = rp_dandelion3.fit_transform(dandelion_data)

dandelion_norm_df1 = np.zeros((len(dandelion_data),len(dandelion_data)))
dandelion_norm_df2 = np.zeros((len(dandelion_data),len(dandelion_data)))
dandelion_norm_df3 = np.zeros((len(dandelion_data),len(dandelion_data)))
for i in range(0,len(dandelion_data)):
    for j in range(0,len(dandelion_data)):
        norm_dif = np.linalg.norm(dandelion_rp1[i]-dandelion_rp1[j])
        dandelion_norm_df1[i,j]=norm_dif
for i in range(0,len(dandelion_data)):
    for j in range(0,len(dandelion_data)):
        norm_dif = np.linalg.norm(dandelion_rp2[i]-dandelion_rp2[j])
        dandelion_norm_df2[i,j]=norm_dif
for i in range(0,len(dandelion_data)):
    for j in range(0,len(dandelion_data)):
        norm_dif = np.linalg.norm(dandelion_rp3[i]-dandelion_rp3[j])
        dandelion_norm_df3[i,j]=norm_dif
        
name1 = 'dandelion_after_proj '+ str(k_dandelion1) +'.csv'
name2 = 'dandelion_after_proj '+ str(k_dandelion2) +'.csv'
name3 = 'dandelion_after_proj '+ str(k_dandelion3) + '.csv'

pd.DataFrame(dandelion_norm_df1).to_csv(name1)
pd.DataFrame(dandelion_norm_df2).to_csv(name2)
pd.DataFrame(dandelion_norm_df3).to_csv(name3)

#case of rose data afer projection
n_samples3, n_features3 = rose_data.shape
k_rose1 = johnson_lindenstrauss_min_dim(n_samples3, eps=eps)
k_rose2 = math.ceil(k_rose1 / 2)
k_rose3 = math.ceil(k_rose1/4)



#compute and record record norm of differences between image vectors for rose data after projection
rp_rose1 = GaussianRandomProjection(n_components=k_rose1, random_state=42)
rose_rp1 = rp_rose1.fit_transform(rose_data)
rp_rose2 = GaussianRandomProjection(n_components=k_rose2, random_state=42)
rose_rp2 = rp_rose2.fit_transform(rose_data)
rp_rose3 = GaussianRandomProjection(n_components=k_rose3, random_state=42)
rose_rp3 = rp_rose3.fit_transform(rose_data)

rose_norm_df1 = np.zeros((len(rose_data),len(rose_data)))
rose_norm_df2 = np.zeros((len(rose_data),len(rose_data)))
rose_norm_df3 = np.zeros((len(rose_data),len(rose_data)))
for i in range(0,len(rose_data)):
    for j in range(0,len(rose_data)):
        norm_dif = np.linalg.norm(rose_rp1[i]-rose_rp1[j])
        rose_norm_df1[i,j]=norm_dif
for i in range(0,len(rose_data)):
    for j in range(0,len(rose_data)):
        norm_dif = np.linalg.norm(rose_rp2[i]-rose_rp2[j])
        rose_norm_df2[i,j]=norm_dif
for i in range(0,len(rose_data)):
    for j in range(0,len(rose_data)):
        norm_dif = np.linalg.norm(rose_rp3[i]-rose_rp3[j])
        rose_norm_df3[i,j]=norm_dif
        
name1 = 'rose_after_proj '+ str(k_rose1) +'.csv'
name2 = 'rose_after_proj '+ str(k_rose2) +'.csv'
name3 = 'rose_after_proj '+ str(k_rose3) + '.csv'

pd.DataFrame(rose_norm_df1).to_csv(name1)
pd.DataFrame(rose_norm_df2).to_csv(name2)
pd.DataFrame(rose_norm_df3).to_csv(name3)

#case of sunflower data
n_samples4, n_features4 = sunflower_data.shape
k_sunflower1 = johnson_lindenstrauss_min_dim(n_samples4, eps=eps)
k_sunflower2 = math.ceil(k_sunflower1 / 2)
k_sunflower3 = math.ceil(k_sunflower1/4)



#compute and record record norm of differences between image vectors for sunflower data after projection
rp_sunflower1 = GaussianRandomProjection(n_components=k_sunflower1, random_state=42)
sunflower_rp1 = rp_sunflower1.fit_transform(sunflower_data)
rp_sunflower2 = GaussianRandomProjection(n_components=k_sunflower2, random_state=42)
sunflower_rp2 = rp_sunflower2.fit_transform(sunflower_data)
rp_sunflower3 = GaussianRandomProjection(n_components=k_sunflower3, random_state=42)
sunflower_rp3 = rp_sunflower3.fit_transform(sunflower_data)

sunflower_norm_df1 = np.zeros((len(sunflower_data),len(sunflower_data)))
sunflower_norm_df2 = np.zeros((len(sunflower_data),len(sunflower_data)))
sunflower_norm_df3 = np.zeros((len(sunflower_data),len(sunflower_data)))
for i in range(0,len(sunflower_data)):
    for j in range(0,len(sunflower_data)):
        norm_dif = np.linalg.norm(sunflower_rp1[i]-sunflower_rp1[j])
        sunflower_norm_df1[i,j]=norm_dif
for i in range(0,len(sunflower_data)):
    for j in range(0,len(sunflower_data)):
        norm_dif = np.linalg.norm(sunflower_rp2[i]-sunflower_rp2[j])
        sunflower_norm_df2[i,j]=norm_dif
for i in range(0,len(sunflower_data)):
    for j in range(0,len(sunflower_data)):
        norm_dif = np.linalg.norm(sunflower_rp3[i]-sunflower_rp3[j])
        sunflower_norm_df3[i,j]=norm_dif
        
name1 = 'sunflower_after_proj '+ str(k_sunflower1) +'.csv'
name2 = 'sunflower_after_proj '+ str(k_sunflower2) +'.csv'
name3 = 'sunflower_after_proj '+ str(k_sunflower3) + '.csv'

pd.DataFrame(sunflower_norm_df1).to_csv(name1)
pd.DataFrame(sunflower_norm_df2).to_csv(name2)
pd.DataFrame(sunflower_norm_df3).to_csv(name3)

#case of tulip data
n_samples5, n_features5 = tulip_data.shape
k_tulip1 = johnson_lindenstrauss_min_dim(n_samples5, eps=eps)
k_tulip2 = math.ceil(k_tulip1 / 2)
k_tulip3 = math.ceil(k_tulip1/4)



#compute and record record norm of differences between image vectors for tulip data after projection
rp_tulip1 = GaussianRandomProjection(n_components=k_tulip1, random_state=42)
tulip_rp1 = rp_tulip1.fit_transform(tulip_data)
rp_tulip2 = GaussianRandomProjection(n_components=k_tulip2, random_state=42)
tulip_rp2 = rp_tulip2.fit_transform(tulip_data)
rp_tulip3 = GaussianRandomProjection(n_components=k_tulip3, random_state=42)
tulip_rp3 = rp_tulip3.fit_transform(tulip_data)

tulip_norm_df1 = np.zeros((len(tulip_data),len(tulip_data)))
tulip_norm_df2 = np.zeros((len(tulip_data),len(tulip_data)))
tulip_norm_df3 = np.zeros((len(tulip_data),len(tulip_data)))
for i in range(0,len(tulip_data)):
    for j in range(0,len(tulip_data)):
        norm_dif = np.linalg.norm(tulip_rp1[i]-tulip_rp1[j])
        tulip_norm_df1[i,j]=norm_dif
for i in range(0,len(tulip_data)):
    for j in range(0,len(tulip_data)):
        norm_dif = np.linalg.norm(tulip_rp2[i]-tulip_rp2[j])
        tulip_norm_df2[i,j]=norm_dif
for i in range(0,len(tulip_data)):
    for j in range(0,len(tulip_data)):
        norm_dif = np.linalg.norm(tulip_rp3[i]-tulip_rp3[j])
        tulip_norm_df3[i,j]=norm_dif
        
name1 = 'tulip_after_proj '+ str(k_tulip1) +'.csv'
name2 = 'tulip_after_proj '+ str(k_tulip2) +'.csv'
name3 = 'tulip_after_proj '+ str(k_tulip3) + '.csv'

pd.DataFrame(tulip_norm_df1).to_csv(name1)
pd.DataFrame(tulip_norm_df2).to_csv(name2)
pd.DataFrame(tulip_norm_df3).to_csv(name3)
```
